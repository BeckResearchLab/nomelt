## make the optimizer trajectory gif with vmd

```python
from nomelt.thermo_estimation.optimizer import OptTrajSuperimposer
import pandas as pd
```

```python
df = pd.read_csv('../data/enh/optimize_enh1_trials.csv')
```

```python
df = df[df['state'] == 'COMPLETE']
```

```python
df = df.sort_values(by='number')
```

```python
df.columns
```

```python
values = list(df['value'].values)
structs = list(df['user_attrs_pdb_file'].values)
seqs = list(df['user_attrs_variant_seq'].values)
```

```python
min(values)
```

```python
structs = [s+'/ensemble_replicate_0.pdb' for s in structs]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(values)
```

```python
movie_maker = OptTrajSuperimposer(sequences=seqs, structure_files = structs, values=values, output_dir='./tmp/opt_traj')
```

```python
movie_maker.run()
```

```python
movie_maker.make_optimization_movie()
```
